```python
def facto(num):
    fac=1
    i=1
    while (i<= num):
        fac = fac*i
        i +=1
        #break
    print('factorial of ', num , 'is ', fac)

facto(4)

squares = [x**2 for x in range(1,6)]
print(squares)
n=[1,2,3,4,5,6]
en = [x for x in n if x%2==0]
print (en)
cd=['even' if x%2==0 else 'odd' for x in n]
print(cd)
total_even = sum([1 for x in cd if x =='odd'])
print ('Total Odd ', total_even)

def dfar(name='dhananjay'):
    print('hello ', name)
dfar('vansh')

# funtion
def add_num(*args): #*args — Variable-length Positional Arguments
    total=0
    for x in args:
        total += x
    return total

print(add_num(2,3))
print(f"add of list items : {add_num(*[1,2,3])}")
#print(add_num(range(5))) this will not work so use unpack * is used for unpacking list, touples, ** is used for unpacking dict

print('unpacking range as a single input', add_num(*range(5)))

def user_info(**kwargs) : #**kwargs — Variable-length Keyword Arguments
    for key,value in kwargs.items():
        print(f"{key}: {value}")

print(user_info(name="Dhananjay", age=35, city="Delhi"))

def show_details(*args, **kwargs): # combined *args and **kwargs
    print("Positional args:", args)
    print("Keyword args:", kwargs)

show_details(10, 20, name="DJ", country="India")

# lambda function is small, oneline, unnamed function defined using labmda
# lambda arguments: expression
square = lambda x : x*x
print(f'square using lambda of 5 is : {square(5)}')

add = lambda a,b : a+b
print (add(5,3))

#map, filter and reduce
a = [1,2,3,4,5]
sq = list(map(lambda x:x**2, a))
print(f'square of list item using lambda, {sq}')
#filter
filtered = list(filter(lambda x: x%2==0, a))
print(f'square of list item using lambda, {filtered}')
#reduce : reduce() combines items into single value
# it comes from functools module in python
from functools import reduce
red_add = reduce(lambda x,y: x+y, a)
print(f'Add of list item using lambda, {red_add}')
my_dict={'name':"Dhananjay", 'age':35, 'city':"Delhi"}
for index, item in enumerate(my_dict.items()):
    print(f'enumerates {index} : {item}')
```

```python
#classes
class Person:
    #constructor
    def __init__(self, name, age):
        self.name=name # self is like this
        self.age=age

    # Method
    def greet(self):
        return f'Hi, I am {self.name}'
 #Property
    @property
    def is_adult(self):
        return self.age >= 18

#Usage
p=Person("Dhananjay", 40)
print(p.greet())
print(f'Are you adult? : {'Yes' if p.is_adult else 'No'}')
        
```

```python
#CSV/JSON 
import csv

with open("marks.csv", mode="r", newline='') as file:
    # crate a reader objectd
    reader = csv.reader(file)
    #iterate over the rows
    index=0
    result=[]
    for row in reader:
        if(index == 0) : # skip as it is column name
             print(row)
        else:
            marks = int(row[2])
            if(marks > 40):
                result.append(row)
                    
        index += 1
        
        #print(row) # each row is a list of strings
    print(f"Student with above 40 marks : {result}")
        
    
```
